```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Simulate data with a linear relationship
np.random.seed(42)
x = np.linspace(0, 10, 100)
y = 3 * x + np.random.normal(0, 2, 100)  # linear relationship with some noise

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Fit an OLS model
model = smf.ols('y ~ x', data=df).fit()

# Predict values
df['y_pred'] = model.predict(df['x'])

# Plot the data and the fitted regression line
plt.scatter(df['x'], df['y'], label='Observed data', color='blue')
plt.plot(df['x'], df['y_pred'], label='Fitted line', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Fitted Regression Line')
plt.legend()
plt.show()

# Print the summary of the model
print(model.summary())
```
